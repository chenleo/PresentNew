### Abstract:

DNA is wrapped around histones(H2A, H2B, H3, and H4) to form nucleosomes. Nucleosomes are further compacted to form condensed chromatin. Molecular Dynamics simulation provide an atomic level describe of nucleosome dynamics. To understand the internal stability, we performed MD simulation and applied force on the end of DNA of nucleosome(with different forces). By using force distribution analysis based on residue level, we can directly view how does force propogate along DNA-histone under tension. The force distribution analysis method is not limited in Nucleosome, but can apply to any kind of MD simulation.

# Material and Method

The nucleosome crystal structure is used 1ZBB. Description of the structure on different scales:

- System:
    - 1 nucleosome
- Molecules:
    - 8 histone subunits
    - 2 DNA single strand
- Residues:
    - 1306 residues:
        - 974 Amino acid for histone
        - 332 Nucleotide for DNA (147 core base pairs and 19 extra handle)
- Atoms:
    - 26305 atoms

Simulation were run with Gromacs-4.5 software package, Force Distribution Analysis program was developed based on Gromacs-5.0dev lib(https://github.com/gromacs/gromacs). Details of Force analysis see [Appendix](#appendix)

Nucleosome was incubate into a solvante box with 100mM NaCl, dimension of 30nmX16.5nmX12nm, to make sure enough space for pulling and distance from mirror nucleosome in periodic boundary condition is larger than 2.0nm. Energy minimization was taken for 2ns, and equilibrate for 30ns. Afterwards, external force was loaded on DNA ends, pull on parallel different direction, with different force value. 

As force was calculated with reduced unit, in the following, we use 1 force unit = 1.66pN without special specification. 

To detail understand the dynamics of nucleosome, we re-calculated force distribution. And do the analysis on residue level.

#Appendix

### 1.Core design:

The force distribution analysis is a plug-in program, developed besided on Gromacs-5.0dev (https://github.com/gromacs/gromacs). No core orginal codes were modified.

During the normal MD simulation, force was calulated by Particle mesh Ewald(PME) method on Fourier space, the total force applied on each atom was summed, and there is no information about where and how does these forces come from. To calculate pair-wise force interaction, the non-bonded force kernal was changed to Coul instead of Ewald.

Input parameters are configured in .mdp file, mainly the short distance cut-off, no extra input file is needed. As non-bonded kernal are used differently, Analyse of simulation trajectories with water molecules are not garanteed.

Force was output based on residue level

Output files:

- summary.csv: 
    - total mass and charge for each residues.
- center_of_mass.csv:
    - center of mass for each residue, full frames
- force_bonded.csv: 
    - Vector bonded force for each residue, full frames
- force_external.csv:
    - Scalar external non-bonded force (Coul and Vdw) based on molecules for each residues, full frames 
- force_internal.csv:
    - Scalar internal non-bonded force (Coul and Vdw) based on molecules for each residues, full frames 
- force_map.csv:
    - Residue-Residue interaction scalar force map, at most 100ps/frame. 

Data analysis and visualisition are all powered by IPython/IPython-notebook<cite data-cite="Ipython">(Granger, 2013)</cite>. Relay on modules:
    - numpy
    - pandas
    - matplotlib

This program is not limited for simulation of nucleosome, it also can be used for any other kind of simulation as well.

### 2.Introduction to MD simulation:

- Simulations are classical:
    - For a system of N interacting <font color="red">atoms</font>:
        - $m_i {\partial^2 \vec{r}_i \over \partial t^2} = \vec{F}_i, i = 1...N$
    - Forces are negative derivatives of a potential function V:
        - $\vec{F}_i = - {\partial V \over \partial \vec{r}_i}$
- Electrons are in the ground state
- Force Fields are approximate
- Pair-additive: (apart from long-range Coulomb forces, approximated by Ewald mesh)
    - $\vec{F}_{ij} = -\vec{F}_{ji}$
- Long-range interactions are cut off
- Boundary conditions are unnatural
    - "Crowded" environment

```python
%%writefile forceAnalysis.py

from __future__ import print_function, division
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

import os

#for nucleosome
#residue range:
H3_1 = np.arange(1,136)          #0
H3_2 = np.arange(488, 623)       #4
H4_1 = np.arange(136,238)        #1
H4_2 = np.arange(623,725)        #5
H2A_1 = np.arange(238,366)       #2
H2A_2 = np.arange(725,853)       #6
H2B_1 = np.arange(366,488)       #3
H2B_2 = np.arange(853,975)       #7
DNA = np.arange(975,1307)        # 8 and 9
DNA_1 = np.arange(975,1141)    #8
DNA_2 = np.arange(1141,1307)   #9
totalRange = np.arange(1,1307) #0 -> 9

#for the external force
listFile = os.listdir(".")
assert "force_map.csv" in listFile , \
                        "There is no file called force_map.csv in this directory"

#Directly read in the summary information
generalInfo = pd.read_csv("summary.csv")

#Get global residue names
globalResidueNames = generalInfo.columns

#Get the total residue number
residueNumber = len(generalInfo.columns[1:])

#for force_external.csv
def readExternal():
    """
    read in force_external.csv
    """
    return pd.read_csv("force_external.csv")

#for force_internal.csv
def readInternal():
    """
    read in force_internal.csv
    """
    return pd.read_csv("force_internal.csv")

#for force_bonded.csv
def readBonded():
    """
    read in force_bonded.csv
    """
    return pd.read_csv("force_bonded.csv")

#for force_map.csv
def readForceMap():
    """
    read in force_map.csv
    """
    print("Warning: force_map.csv can be very large.")
    return pd.read_csv("force_map.csv")

#for center of mass
def readCenterMass():
    """
    read in center_of_mass.csv
    """
    return pd.read_csv("center_of_mass.csv")

#get range 
def getRange(df, rangeToAna):
    """
    This function works for force internal and external
    return the rows according to rangeToAna
    """
    assert df.columns.size == generalInfo.columns.size
    return df.irow(rangeToAna)

def getRangeVector(df, rangeToAna):
    """
    This function works for forceBonded and center of mass
    return the rows according to rangeToAna
    """
    assert df.columns.size == (generalInfo.columns.size * 3 - 2)
    return df.irow(rangeToAna)

def getGroup(df, group):
    """
    This function works for force internal and external
    return the columns of group from df
    """
    assert df.columns.size == generalInfo.columns.size
    return df[group]

def getGroupVector(df, group):
    """
    This function works for forceBonded and center of mass
    return the columns of group from df
    """
    assert df.columns.size == (generalInfo.columns.size * 3 - 2)
    rg = group * 3
    return df[np.arange(rg.min()-2, rg.max()+1)]

def getDistance(df, res1, res2):
    """
    This function returns the center of mass distance from center_of_mas frame
    Between res1 and res2
    res1 and res2 should be global residue index
    """
    assert df.columns.size == (generalInfo.columns.size * 3 - 2)
    traj1 = getGroupVector(df, np.array(res1))
    traj2 = getGroupVector(df, np.array(res2))
    diff = traj1.values - traj2.values
    return np.sqrt((diff*diff).sum(1))
    

#get frame for force map
def getFrame(forceMapFrame, n):
    """
    read the n th frame from forceMapFrame
    return the frame and time(ps)
    """
    assert len(forceMapFrame) % residueNumber == 0
    assert len(forceMapFrame) / residueNumber >= n
    assert n >= 0
    return forceMapFrame.irow(np.arange(residueNumber) + n*residueNumber), n*100

#get interaction map between group1 and group2 from certain force frame
def getInteraction(frame, group1, group2):
    """
    Please note: The head index shows the list of group2
    """
    assert len(frame) == residueNumber
    assert frame.columns.size - 1 == residueNumber 
    return frame.irow(group1)[group2]       

#get the force for certain residue    
def getResidueForce(forceMapFrame, n):
    """
    return the force map for certain residue.
    n is the global residue number
    """
    assert len(forceMapFrame) % residueNumber == 0
    return forceMapFrame.irow(np.arange(0,len(forceMapFrame), residueNumber)+n-1)

def getResidueGroup(resFrame, group):
    """
    return the residue force map with group 
    """
    assert resFrame.columns.size -1 ==residueNumber
    return resFrame[group]

#plot heat map:
def plotHeatMap(frame,group1=None, group2=None):
    plt.pcolor(frame, cmap = cm.RdBu_r, vmax=200, vmin=-200)
    plt.colorbar()
    if group2 is not None and len(group2) > 5:
        lenG2 = len(group2)
        plt.xticks(np.arange(0,len(group2),int(lenG2/5)),globalResidueNames[group2[np.arange(0,len(group2),int(lenG2/5))]],
                   rotation='vertical')
    if group1 is not None and len(group1) > 5:
        lenG1 = len(group1)
        plt.yticks(np.arange(0,len(group1),int(lenG1/5)),globalResidueNames[group1[np.arange(0,len(group1),int(lenG1/5))]])
    #plt.xticks(np.arange(0,len(group2),5)
    pass

#Pipline
#Plot group-group interaction for certain frame
def plotCertainFrame(forceFrame, group1, group2, n):
    frameN,tm = getFrame(forceFrame, n)
    intractFrame = getInteraction(frameN, group1, group2)
    #plotHeatMap(intractFrame)
    plotHeatMap(intractFrame, group1=group1,group2=group2)
```
